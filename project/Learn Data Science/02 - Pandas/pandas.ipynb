# Pandas


This short tutorial will take you through the very basics of the Pandas library, a widely used library amongst data scientist to do basic data analysis and operations in. 

Pandas is a Python package providing fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both easy and intuitive. It aims to be the fundamental high-level building block for doing practical, real world data analysis in Python.

The core data structure pandas is the dataframe. It is well suited for many different kinds of data:
- Tabular data with heterogeneously-typed columns, as in an SQL table or Excel spreadsheet
- Ordered and unordered (not necessarily fixed-frequency) time series data.
- Arbitrary matrix data (homogeneously typed or heterogeneous) with row and column labels
- Any other form of observational / statistical data sets. The data actually need not be labeled at all to be placed into a pandas data structure

In this (short!) introduction to pandas we will introduce the basic funtionalities of pandas which you will find useful on a day to day basis as a data scientist.

## A bit of data cleaning

It looks like the top row of the dataframe is wrong. How could we fix it? Pandas dataframes can be sliced, much like python lists. Thus if we wanted to show all the rows of `nfl_data` except the first one, we can simply type

```python
nfl_data[1:]
```

However, this does not change the contant of `nfl_data` itself:

```python
nfl_data.head(5)
```

This is because we did not assign anything to `nfl_data`. This is how we do it:

```python
nfl_data = nfl_data[1:]
```

```python
nfl_data.head(5)
```

To get some basic information about the columns of our dataframe, we can use the `info()` member function of the dataframe:

```python
nfl_data.info()
```

Each row above represents a column of the dataframe, and to the right of the column name the type of data it contains is indicated. In this case, each column contains data of of type `object` which in `pandas` means that they are treated as strings. Normally, the `pandas.read_csv()` will automatically infer the type of data it reads from a csv file, for example treating `54` as an integer rather than a string. However, in this case it seems to have failed to do so. Why do you think that is?

Since the type of the columns is currently incorrect, let us fix it. For example, let us tell `pandas` to explicitly treat the column `Passing Attempts` as a column of integers.

```python
nfl_data['Passing Attempts'] = pd.to_numeric(nfl_data['Passing Attempts'])
```

Let's check if it worked.

```python
nfl_data.info()
```

As you can see, the type for the column `Passing Attempts` is now `int64`, which means that it is being treated as a (64-bit) integer. What about the column `Rushing Yards`? 


```python
nfl_data['Rushing Yards'] = pd.to_numeric(nfl_data['Rushing yards'])
```

Running the above cell will cause an error. You will see this error a lot! It means that the column name you entered does not correspond to any column in the dataframe. In fact, you will notice that many columns have a leading blank space character in their name. Let's try and remove them. The column names can be accessed as follows:

```python
nfl_data.columns
```

To  remove the spaces, we can do the following:

```python
nfl_data.columns = [column_name.strip(' ') for column_name in nfl_data.columns]
```

Let's have another look at the columns:

```python
nfl_data.info()
```

This looks much better. However, half the columns have a `.1` suffix, which is rather confusing. These correspond correspond to the statistics of the home team. So let's replace every `.1` suffix to `_H`, which is more self-explanatory.



```python
nfl_data.columns = [x.replace('.1', '_H') for x in nfl_data.columns]
```

How would you append `_H` to the name of each column corresponding to the statistics of the visiting team? 

```python
#Answer
nfl_data.columns = [x + '_V' 
                    if x.find('_H')==-1 and x not in ['Date', 'Vis Team', 'Home Team'] 
                    else x
                    for x in nfl_data.columns]
```

Now that we are happy with the column names, why don't we save the dataframe to a csv file? This can be done by using the `to_csv()` member function of `nfl_data`.

```python
nfl_data.to_csv('cleaned_nfl_data.csv',index=False)
```

The argument `index=False` that we passed on to `to_csv()` indicates that we do not wish to store the index of the dataframe to the csv file. Let's reload our cleaned dataframe from the csv file and see if the column types have changed.

```python
nfl_data = pd.read_csv('cleaned_nfl_data.csv')
```

```python
nfl_data.info()
```

As you can see, most of the columns are correctly identified as containing integers. As you will have guessed by now, the reason why these types were not correctly identified in the first place is that the header row of the csv file, which contains the names of the columns was duplicated, to the effect that `read_csv()` defaulted to the basic `object` type since the column contained conflicting types. Notice, however, that the `Date` column is still incorrectly identified. Indeed, `pandas` has a special type for dates. Let's convert this column to the appropriate format.

```python
nfl_data['Date'] = pd.to_datetime(nfl_data['Date'])
```

Let's introduce one last useful `pandas` funtion. The `describe()` method computes a few important statistics for each column of the dataframe.

```python
nfl_data.describe()
```

# More about pandas


How could we find out how many teams there are in the NFL? Well, we know they all must play at home, so we could print them all:

```python
nfl_data['Home Team']
```

But what we really want is a just each team once. This is a set!

```python
set(nfl_data['Home Team'])
```

And we could manually count! or we could use len()

```python
len(set(nfl_data['Home Team']))
```

However, `pandas` already provides a neat function for this, namely the `nunique()` function:

```python
nfl_data['Home Team'].nunique()
```

The sister function of `nunique()` is `unique()` which simply lists the unique values appearing in a given column.

```python
nfl_data['Home Team'].unique()
```

We might now want to look at how the teams did at home. Pandas has a great command for this called `groupby()`, which is slightly complex, but which is intuitively easy to use. Let's tell it to group by the home team, and then to sum the columns (computing mean is also possible).

```python
nfl_data.groupby('Home Team').sum()
```

The `groupby()` function is typically used in conjunction with the `apply()` function. The idea is to form group using `groupby()` and then use `apply()` to perform some operation on each group. This is a really handy and powerful feature of `pandas`. In fact, 
> `nfl_data.groupby('Home Team').sum()` 

is a shorthand for

> `nfl_data.groupby('Home Team').apply(lambda x: x.sum())`

as you can see below:

```python
nfl_data.groupby('Home Team').apply(lambda x: x.sum())
```

The `Home Team` column contains leading white blank spaces. Can you use the `apply()` function to get rid of them? 

```python
#Answer
nfl_data['Home Team'] = nfl_data['Home Team'].apply(lambda x: x.strip(' '))
```

How would you return the number of passing completions for the first home game played by each team? You may want to use `sort_values()` to make sure that your data is chronologically ordered.

```python
#Answer
nfl_data.groupby('Home Team').apply(lambda x: x.sort_values(['Date'])['Passing Completions_H'].head(1))
```
